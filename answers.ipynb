## Euler Method

Equation 4.8 from Pang: $y_{i+1}=y_{i}+\tau g_{i}$, where g is 

```python
def euler(x0,v0,t0,t_final,time_step):
    t = np.arange(t0,t_final,time_step)
    k = 1.
    m = 1.
    x = np.zeros(len(t))
    v = np.zeros(len(t))
    x[0] = x0
    v[0] = v0
    i=0
    while i<len(t)-1:
        x[i+1] = x[i]+v[i]*(time_step)
        v[i+1] = v[i]+(-k/m)*x[i]*time_step
        
        i+=1
    return x,v,t

def corrector(x0,v0,t0,t_final,time_step):
    t = np.arange(t0,t_final,time_step)
    k = 1.
    m = 1.
    x = np.zeros(len(t))
    v = np.zeros(len(t))
    x[0] = x0
    v[0] = v0
    i=0
    while i<len(t)-1:
        x[i+1] = x[i]+v[i]*(time_step)
        v[i+1] = v[i]+(-k/m)*x[i]*time_step
        i+=1
    while i<len(t)-1:
        x[i+1] = x[i]+(v[i]+v[i+1])*time_step/2
        v[i+1] = v[i]-(x[i]+x[i+1])*time_step/2
        i+=1
    return x,v,t
```

```python
x,v,t = euler(0,1,0,4*np.pi,0.1*np.pi)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.subplot()
av = plt.subplot()

ax.plot(t,x,'--',color='blue',label='position')
av.plot(t,v,'--',color='red',label='velocity')

plt.xlabel('Time',fontsize=20)
plt.title('Timestep = 0.1$\pi$',fontsize=20)
ax.legend(loc='best')
```
